```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Carregar os dados
df = pd.read_csv('datasets\\SIOE 2024-01-31-utf8.csv', encoding='utf-8')  # Verifica se o caminho está correto
df['NIPC'] = ''  # Adicionar uma nova coluna para armazenar os NIPCs

# Configurar o Selenium
options = webdriver.ChromeOptions()  # Adiciona opções ao navegador se necessário
driver = webdriver.Chrome(options=options)  # Ou outro navegador que estiveres a usar
driver.get("https://www.sioe.dgaep.gov.pt/Default.aspx")

# Aguarda que a página carregue completamente
WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

for index, row in df.iterrows():
    try:
        # Esperar pelo botão para iniciar a pesquisa e clicar nele
        botao_pesquisar = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.ID, "PaginaInicial_btCliqueAqui"))
        )
        botao_pesquisar.click()

        # Esperar pelo campo de designação e inserir o valor
        campo_designacao = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.ID, "Pesquisa_txtDesignacao"))
        )
        campo_designacao.clear()  # Limpar qualquer texto existente antes de digitar um novo
        campo_designacao.send_keys(row['designação'])

        # Encontrar e clicar no botão de pesquisa
        botao_pesquisa = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.ID, "Pesquisa_btnPesquisar"))
        )
        botao_pesquisa.click()

        # Esperar até que o link do resultado seja clicável e clicar
        link_instituto = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "a[id^='Listagem_gvResultados_lnkb_']"))  # Modificado para seleção CSS mais genérica
        )
        link_instituto.click()

        # Esperar até que o elemento NIPC esteja presente e extrair o texto
        nipc_element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.ID, "Detalhes_Geral_Identificacao_Nipc"))
        )
        df.at[index, 'NIPC'] = nipc_element.text  # Armazenar o NIPC extraído na coluna 'NIPC'

        print(f"NIPC para {row['designação']}: {nipc_element.text}")  # Imprimir o NIPC extraído

        # Encontrar e clicar no botão para voltar à pesquisa
        driver.get("https://www.sioe.dgaep.gov.pt/Default.aspx")  # Mudança para recarregar a página e evitar problemas de estado da sessão
    except Exception as e:
        print(f"Erro ao processar {row['designação']}: {e}")

# Fechar o navegador quando terminar
driver.quit()

# Salvar os resultados em um novo arquivo CSV
df.to_csv('resultados_nipc.csv', index=False, encoding='utf-8')
```
